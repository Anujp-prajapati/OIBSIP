```python
x=data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,-1:]
```

```python
xtrain,xtest,ytrain,ytest  = train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
print(x.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
xtrain,ytrain
```

```python
xtest,ytest
```

```python
model = LinearRegression()
```

```python
# Train the model
model.fit(xtrain, ytrain)
```

```python
# Make predictions on the test set
predictions = model.predict(xtest)
predictions
```

```python
model.score(xtrain,ytrain)*100
```

```python
model.score(xtest,ytest)*100
```

```python
mse = mean_squared_error(ytest, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
plt.scatter(xtest['TV'], ytest, color='black', label='Actual')
plt.scatter(xtest['TV'], predictions, color='blue', label='Predicted')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
mean_absolute_error(ytest,predictions)
```

```python
r2_score(ytest,predictions)*100
```

```python
mean = np.sqrt(mean_squared_error(ytest,predictions))
mean
```

```python
cv = KFold(n_splits=5,shuffle=True, random_state=0)
cv
```

```python
score=cross_val_score(model,x,y,cv=cv,n_jobs=-1)
finalscore=np.mean(score)
finalscore
```
